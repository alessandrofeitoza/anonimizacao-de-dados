```python
import pandas as pd
import numpy as np

from pandasgui import show
```

```python
df_covid = pd.read_csv("data/covid.csv", delimiter=";")
df_covid
```

```python
df_regioes = pd.read_csv("data/regioes.csv")
df_regioes
```

# Item 1

```python
def t_municipio(data, nivel=0):
    if nivel == 1:
        return df_regioes[df_regioes['Código IBGE do Município'] == int(data)].values[0][2]
    else:
        return data
```

```python
anon1 = df_covid.copy()
anon1['Municipio'] = anon1['CodigoMunicipio'].apply(lambda x: t_municipio(x, nivel=1))
anon1.to_csv("export/covid_anon01.csv")
```

# Item 2

```python
def t_nascimento(data, nivel=0):
    s = data.split("-")
    if nivel == 1:
        return '-'.join(s[:2])
    elif nivel == 2:
        return '-'.join(s[:1])
    else:
        return data
```

```python
anon2 = df_covid.copy()
anon2['Nascimento'] = anon2['Nascimento'].apply(lambda x: t_nascimento(x, nivel=1))
anon2.to_csv("export/covid_anon02.csv")
```

# Item 3

```python
def fifty_decade(x):
    value = t_nascimento(x, nivel=2)
    if int(value) >= 1950 and int(value) <= 1959:
        return value
    else:
        return x
```

```python
anon3 = df_covid.copy()
anon3['Nascimento'] = anon3['Nascimento'].apply(lambda x: fifty_decade(x))
anon3.to_csv("export/covid_anon03.csv")
```

# Item 4

```python
anon4 = df_covid.copy()
anon4['Municipio'] = anon4['CodigoMunicipio'].apply(lambda x: t_municipio(x, nivel=1))
anon4['Nascimento'] = anon4['Nascimento'].apply(lambda x: t_nascimento(x, nivel=2))
anon4.to_csv("export/covid_anon04.csv")
```
